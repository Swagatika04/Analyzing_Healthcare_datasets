```python
#importing libraries
import pandas as pd
import numpy as np
import datetime 
```

```python
#reading csv tables
df_visit = pd.read_csv(r"C:\Users\sagat\OneDrive\Documents\healthcare\visit.csv")
df_diagnosis = pd.read_csv(r"C:\Users\sagat\OneDrive\Documents\healthcare\diagnosis.csv")
df_department = pd.read_csv(r"C:\Users\sagat\OneDrive\Documents\healthcare\department.csv")
df_medication_order = pd.read_csv(r"C:\Users\sagat\OneDrive\Documents\healthcare\medication_order.csv")
df_visit_diagnosis = pd.read_csv(r"C:\Users\sagat\OneDrive\Documents\healthcare\visit_diagnosis.csv")
```

```python
#Hospital encounter filter
df_visit_hospital = df_visit[df_visit["DICT_ENC_TYPE_KEY"]== 83]
```

```python
#info of the new dataset
df_visit_hospital.info()
```

```python
#Changing Hospital encounter unique key to datetime and filter the dataframe by date
df_visit_hospital["CONTACT_DT_KEY"]= pd.to_datetime(df_visit_hospital["CONTACT_DT_KEY"],format='%Y%m%d')
df_visit_hospital_date = df_visit_hospital[df_visit_hospital["CONTACT_DT_KEY"] >="2014/08/01"]
```

```python
#looking at the first few rows
df_visit_hospital_date.head()
```

```python
# looking at the filtered dataset
df_visit_hospital_date.info()
```

```python
#AGE filter
df_visit_hospital_date_age = df_visit_hospital_date[(df_visit_hospital_date.AG >1) & (df_visit_hospital_date.AGE < 18) ]
```

```python
#Merging visit diagnosis and diagnosis file, and filtering by ICD9_CD code and emergency codes
df_merge = pd.merge(df_visit_diagnosis, df_diagnosis, on = 'DX_KEY' )
ICD9_code = ['995.0' ,'995.3','995.6','995.60','995.61','995.62','995.63','995.64','995.65','995.66','995.67','995.68','995.69','995.7','999.4','999.41','999.42','999.49']
df_visit_diagnosis_emergency=df_merge[(df_merge.DICT_DX_STS_KEY.isin(["313", "314"]))& (df_merge.ICD9_CD.isin(ICD9_code)) ]
df_visit_hospital_date_age_emergency=df_visit_hospital_date_age[df_visit_hospital_date_age.VISIT_KEY.isin(df_visit_diagnosis_emergency.VISIT_KEY)]
```

```python
# Checking the new datasets
df_visit_hospital_date_age_emergency.info()
```

```python
#Filering for encounters not at Urgent Care Dept
df_department_filtered= df_department[df_department.DEPT_CNTR.str.contains('URGENT CARE') == False]
df_patient_cohort=df_visit_hospital_date_age_emergency[df_visit_hospital_date_age_emergency.DEPT_KEY.isin(df_department_filtered.DEPT_KEY)]
```

```python
df_patient_cohort.info()
```

```python
#New column created and given value 1 if diagnosis text contains "anaphylaxis"
df_merge_ana= df_merge[df_merge['DX_NM'].str.contains('anaphylaxis', case=False)]
df_patient_cohort['ANAPH_DX_IND'] = np.where(df_patient_cohort['VISIT_KEY'].isin(df_merge_ana.VISIT_KEY), 1, 0)
```

```python
#New column created and given value 1 if medication order name contains "epinephrine"
df_medication_order=df_medication_order.dropna(subset=['MED_ORD_NM'])
df_medication_order_epi= df_medication_order[df_medication_order['MED_ORD_NM'].str.contains('epinephrine',case=False)]
df_patient_cohort['EPI_ORDER_IND'] = np.where(df_patient_cohort['VISIT_KEY'].isin(df_medication_order_epi.VISIT_KEY), 1, 0)
```

```python
df_patient_cohort.head()
```

```python
#Dropping columns not required and converting datetime columns to datetime
df_patients_cohort_filter = df_patient_cohort[['PAT_KEY','VISIT_KEY','HOSP_ADMIT_DT','HOSP_DISCHRG_DT','AGE','ANAPH_DX_IND','EPI_ORDER_IND']]
df_patients_cohort_filter["HOSP_DISCHRG_DT"]=pd.to_datetime(df_patients_cohort_filter["HOSP_DISCHRG_DT"])
df_patients_cohort_filter["HOSP_ADMIT_DT"]=pd.to_datetime(df_patients_cohort_filter["HOSP_ADMIT_DT"])
```

```python
df_patients_cohort_filter.head()
```

```python
#Office vist encounter dataframe
df_office=df_visit[df_visit["DICT_ENC_TYPE_KEY"]==108]
df_office_filter= df_office[['PAT_KEY','APPT_CHECKIN_DT']]
```

```python
#Merging patient cohort with office visit df on patient key and with a left join
df_office_hospital = pd.merge(df_patients_cohort_filter, df_office_filter, on = 'PAT_KEY', how="inner")
df_office_hospital=df_office_hospital[['PAT_KEY','VISIT_KEY','HOSP_DISCHRG_DT','APPT_CHECKIN_DT']]
```

```python
#Converting datetime columns to datetime
df_office_hospital["HOSP_DISCHRG_DT"]=pd.to_datetime(df_office_hospital["HOSP_DISCHRG_DT"])
df_office_hospital["APPT_CHECKIN_DT"]=pd.to_datetime(df_office_hospital["APPT_CHECKIN_DT"])
```

```python
#follow up indicator is 1 if follow up with 7 days
df_office_hospital["FOLLOW_UP_IND"]=np.where(((df_office_hospital.APPT_CHECKIN_DT.dt.date-df_office_hospital.HOSP_DISCHRG_DT.dt.date)/ (np.timedelta64(1, 'D'))).astype(int).isin([0,1,2,3,4,5,6,7]), 1, 0)
df_office_hospital_follow=df_office_hospital[df_office_hospital.FOLLOW_UP_IND==1]
```

```python
#Days to follow up calculated if with 7 days and follow up column added
df_office_hospital_follow["DAYS_TO_FOLLOW_UP"]=((df_office_hospital_follow.APPT_CHECKIN_DT.dt.date-df_office_hospital_follow.HOSP_DISCHRG_DT.dt.date)/ (np.timedelta64(1, 'D'))).astype(int)
df_office_hospital_follow["FOLLOW_UP_DATE"]=df_office_hospital_follow["APPT_CHECKIN_DT"]
df_follow_final=df_office_hospital_follow.loc[df_office_hospital_follow.groupby(["VISIT_KEY"])["DAYS_TO_FOLLOW_UP"].idxmin()].reset_index()
```

```python
#Merging newly created columns to patient cohort with a left join
patient_cohort_final = pd.merge(df_patients_cohort_filter, df_follow_final, on = ["VISIT_KEY", "PAT_KEY", "HOSP_DISCHRG_DT"], how = "left")
```

```python
#Dropping column not needed in final table
patient_cohort_final= patient_cohort_final.drop(["HOSP_DISCHRG_DT", "index", "APPT_CHECKIN_DT"], axis=1)
```

```python
#Converting columns to categorical
patient_cohort_final.ANAPH_DX_IND= pd.Categorical(patient_cohort_final.ANAPH_DX_IND)
patient_cohort_final.EPI_ORDER_IND= pd.Categorical(patient_cohort_final.EPI_ORDER_IND)
patient_cohort_final.FOLLOW_UP_IND= patient_cohort_final.FOLLOW_UP_IND.fillna(0)
patient_cohort_final.FOLLOW_UP_IND =pd.Categorical(patient_cohort_final.FOLLOW_UP_IND)
```

```python
#Exporting dateframe to csv
patient_cohort_final.to_csv("~/Desktop/patient_cohort.csv")
```
